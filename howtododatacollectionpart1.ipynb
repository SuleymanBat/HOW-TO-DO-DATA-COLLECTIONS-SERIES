# Using indexing, retrieve the string 'willow' from the list and assign that to the variable plant.

```python
data = ['bagel','cream cheese','breakfast','grits','eggs','bacon',[34,9,73,[]],[['willow','birch','elm'],'apple','peach','cherry']]
plant = data[7]
print(plant)
```

```python
data = ['bagel','cream cheese','breakfast','grits','eggs','bacon',[34,9,73,[]],[['willow','birch','elm'],'apple','peach','cherry']]
plant = data[7][0]
print(plant)
```

```python
data = ['bagel','cream cheese','breakfast','grits','eggs','bacon',[34,9,73,[]],[['willow','birch','elm'],'apple','peach','cherry']]
plant = data[7][0][0]
print(plant)
```

# NESTED DICTIONARIES

## TASK: 
Extract the value associated with the key color and assign it to the variable color.

```python
info = {'personal_data':
       {'name':'Lauren',
       'age':20,
       'major':'Information Science',
       'physical_features':
       {'color':{'eye':'blue',
                'hair':'brown'},
       'height':"5'8"}
       },
       'other':
       {'favorite_colors': ['purple','green','blue'],
       'interested_in': ['social media','intellectual property','copyright','music','books']
       }
       }
```

```python
color = info['personal_data']
print(color)
```

```python
color = info['personal_data']['physical_features']
print(color)
```

```python

color = info['personal_data']['physical_features']['color']
print(color)
```

# Suppose we wanted to change the value that was associated with a color key

```python
info['personal_data']['physical_features']['color'] = 95
color = info['personal_data']['physical_features']['color']
print(color)
```
